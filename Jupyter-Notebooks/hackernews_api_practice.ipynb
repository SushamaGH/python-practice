```python
import requests

url="https://hacker-news.firebaseio.com/v0/maxitem.json"

baseurl="https://hacker-news.firebaseio.com/v0/item/"
data = []
request = requests.get(url)
max_id = request.json()


for i in range(100):
    url = baseurl + str(max_id) + ".json"
    max_id  = max_id - 1
    url = baseurl + str(max_id) + ".json"
    print(url)
    try:
        request = requests.get(url)
        response = request.json()
        data.append(response)
    except:
        print("Story does not exist anymore!")
        continue
    
```

```python
for d in data:
    print(d)
    
```

```python
#2. Write a function to display all the items written by a certain user
def items_by_user(user):
    for d in data:
        if "by" in d:
            if d["by"] == user:
               return d["id"]
print(items_by_user("Genbox"))
```

```python
#3. Write a function to display the average score of all items
def avg_score_all_items():
    total = 0
    score = 0
    for d in data:
        if "score" in d:
            total += 1
            score += d["score"]
    avg = round(score/total, 2)
    return avg

print(avg_score_all_items())

```
